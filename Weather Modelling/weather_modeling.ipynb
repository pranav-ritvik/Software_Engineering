# Implement weather modeling using the quadratic solution with hardcoding variables
# Single input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Hard-coded coefficients
a = 1.0
b = -2.0
c = 3.0

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Calculate temperature values using the quadratic model
temperature_values = quadratic_temperature_model(time_values, a, b, c)

# Plotting the results
plt.plot(time_values, temperature_values, label='Temperature Model')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with hardcoding variables
# Multiple input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Hard-coded sets of coefficients
coefficients = [
    (1.0, -2.0, 3.0),
    (0.5, -1.0, 2.0),
    (1.5, -3.0, 4.0)
]

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for a, b, c in coefficients:
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Temperature Model (a={a}, b={b}, c={c})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with input from keyboard
# Multiple input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Get number of sets
num_sets = int(input("Enter the number of sets of coefficients: "))

# Get user input for multiple sets of coefficients
coefficients = []
for i in range(num_sets):
    print(f"Set {i+1}:")
    a = float(input("Enter the coefficient for the quadratic term (a): "))
    b = float(input("Enter the coefficient for the linear term (b): "))
    c = float(input("Enter the constant term (c): "))
    coefficients.append((a, b, c))

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for a, b, c in coefficients:
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Temperature Model (a={a}, b={b}, c={c})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with input from reading from file
# Single input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Read coefficients from file
with open('coefficients.txt', 'r') as file:
    a, b, c = map(float, file.readline().split())

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Calculate temperature values using the quadratic model
temperature_values = quadratic_temperature_model(time_values, a, b, c)

# Plotting the results
plt.plot(time_values, temperature_values, label='Temperature Model')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with input from reading from file
# Multiple input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Read coefficients from file
coefficients = []
with open('coefficients_multiple_sets.txt', 'r') as file:
    for line in file:
        a, b, c = map(float, line.split())
        coefficients.append((a, b, c))

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for a, b, c in coefficients:
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Temperature Model (a={a}, b={b}, c={c})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```
