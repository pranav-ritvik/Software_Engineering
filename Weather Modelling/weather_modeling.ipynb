# Implement weather modeling using the quadratic solution with hardcoding variables
# Single input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Hard-coded coefficients
a = 1.0
b = -2.0
c = 3.0

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Calculate temperature values using the quadratic model
temperature_values = quadratic_temperature_model(time_values, a, b, c)

# Plotting the results
plt.plot(time_values, temperature_values, label='Temperature Model')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with hardcoding variables
# Multiple input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Hard-coded sets of coefficients
coefficients = [
    (1.0, -2.0, 3.0),
    (0.5, -1.0, 2.0),
    (1.5, -3.0, 4.0)
]

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for a, b, c in coefficients:
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Temperature Model (a={a}, b={b}, c={c})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with input from keyboard
# Multiple input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Get number of sets
num_sets = int(input("Enter the number of sets of coefficients: "))

# Get user input for multiple sets of coefficients
coefficients = []
for i in range(num_sets):
    print(f"Set {i+1}:")
    a = float(input("Enter the coefficient for the quadratic term (a): "))
    b = float(input("Enter the coefficient for the linear term (b): "))
    c = float(input("Enter the constant term (c): "))
    coefficients.append((a, b, c))

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for a, b, c in coefficients:
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Temperature Model (a={a}, b={b}, c={c})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with input from reading from file
# Multiple input set

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Read coefficients from file
coefficients = []
with open('coefficients_multiple_sets.txt', 'r') as file:
    for line in file:
        a, b, c = map(float, line.split())
        coefficients.append((a, b, c))

# Generate time values from 0 to 50 with step 1
time_values = np.arange(0, 51, 1)

# Plotting the results for each set of coefficients
for a, b, c in coefficients:
    temperature_values = quadratic_temperature_model(time_values, a, b, c)
    plt.plot(time_values, temperature_values, label=f'Temperature Model (a={a}, b={b}, c={c})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature Model')
plt.legend()
plt.grid(True)
plt.show()
```

# Implement weather modeling using the quadratic solution with input from data set


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def quadratic_temperature_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

# Read dataset from CSV file using pandas
data = pd.read_csv('data_set.csv')

# Extract columns from the dataframe
a_values = data['a']
b_values = data['b']
c_values = data['c']
time_columns = ['time0', 'time1', 'time2', 'time3', 'time4', 'time5']
temp_columns = ['temp0', 'temp1', 'temp2', 'temp3', 'temp4', 'temp5']

# Plotting the results for each set of coefficients
for index, row in data.iterrows():
    a = row['a']
    b = row['b']
    c = row['c']
    time_values = row[time_columns].values
    observed_temps = row[temp_columns].values
    
    computed_temps = quadratic_temperature_model(time_values, a, b, c)
    
    plt.figure(figsize=(10, 5))
    plt.plot(time_values, computed_temps, label='Computed Temperature', linestyle='--')
    plt.plot(time_values, observed_temps, label='Observed Temperature', marker='o')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title(f'Quadratic Temperature Model (a={a}, b={b}, c={c})')
    plt.legend()
    plt.grid(True)
    plt.show()
```
