```python
import matplotlib.pyplot as plt
```

```python
# 1. Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.grid(True)
plt.show()
```

```python
# 2. Scatter Plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='red', marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# 3. Histogram
data = [1, 1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 7, 8, 9]

plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True)
plt.show()
```

```python
# 4. Bar Plot
labels = ['A', 'B', 'C', 'D']
values = [10, 30, 20, 15]

plt.bar(labels, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.grid(True)
plt.show()
```

```python
# 5. Pie Chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 25, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```
